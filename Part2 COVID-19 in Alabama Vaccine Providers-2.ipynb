```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
import warnings


%matplotlib inline 
```

```python
file = pd.read_excel("C:/Users/dassi/Downloads/COVID-19 in Alabama Vaccine Providers-2.xlsx")
file
```

```python
file.describe()
```

```python
# Group by the 'County' column and count the number of vaccine providers per county
county_provider_counts = file.groupby('County').size().reset_index(name='Number of Providers')

# Sort by the number of providers in descending order
county_provider_counts = county_provider_counts.sort_values(by='Number of Providers', ascending=False)

# Display the result
print(county_provider_counts)

# Optional: Plot the results using a bar chart for a better visual representation
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar plot
plt.figure(figsize=(12, 13))
sns.barplot(x='Number of Providers', y='County', data=county_provider_counts, palette='viridis')
plt.title('Number of Vaccine Providers by County in Alabama')
plt.xlabel('Number of Providers')
plt.ylabel('County')
plt.show()
```
